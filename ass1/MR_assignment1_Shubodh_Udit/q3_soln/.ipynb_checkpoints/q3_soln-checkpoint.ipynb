# Solution to Q3.

1. Given: Points in image and their corresponding points in world. 
2. Aim: Homography matrix `H` which directly relates points in world frame to points in image. It is a combination of `K` matrix(intrinsic parameters), `R` matrix(rotation) and  `t` vector (translation).
3. Below are the equations derived to obtained homography `H` matrix.  




```python
import numpy as np

t = 0.1315; d = 0.0790

# Loading given points
u = [284.56243896, 373.93179321, 387.53588867, 281.29962158, 428.86453247, 524.76373291, 568.3659668, 453.60995483]
v = [149.2925415, 128.26719666, 220.2270813, 241.72782898, 114.50731659, 92.09218597, 180.55757141, 205.22370911]
x = [0, t, t, 0, t+d, 2*t + d, 2*t + d, t+d]
y = [t, t, 0, 0, t, t, 0, 0]

# Storing all the points in world frame into single matrix X
X = np.zeros((3, 8))
for i in range(len(x)):
    X[0:3, i] = [x[i], y[i], 1]

# Generating matrix M which, whose SVD we need to calculate
M = np.zeros((2*len(u), 9))
for i in range(0, 2*len(u), 2):
    p = int(np.floor(i/2))

    M[i, 0:3] = -X[0:3, p]; M[i, 6:9] = u[p]*X[0:3, p]
    M[i+1, 3:6] = -X[0:3, p]; M[i+1, 6:9] = v[p]*X[0:3, p]

# Getting Homography matrix H, from V.T last row
(U, S, Vh) = np.linalg.svd(M, full_matrices=True)
h = Vh[-1,:]
H = h.reshape(3,3)

H = H/H[2, 2]

# Projecting all points to image
X_project = np.dot(H, X)
X_project[0, :] = X_project[0, :]/X_project[2, :]
X_project[1, :] = X_project[1, :]/X_project[2, :]
X_project[2, :] = X_project[2, :]/X_project[2, :]

# Calculating reprojection error
error = 0.0
for i in range(len(u)):
    error += ((X_project[0, i] - u[i])**2 + (X_project[1, i] - v[i])**2)**(0.5) 
print("Reprojection error is: %f pixels" % error)
```

1. As reprojection error is just 2 pixels for all the points so our caculated homography matrix is correct.
2. Now below are projecting world points to image, using calculated homography matrix

```python
# Projecting to image

%pylab inline
import matplotlib.image as mpimg
import matplotlib.pyplot as plt


img=mpimg.imread('./image.png')
imgplot = plt.imshow(img)

plt.plot(X_project[0,:], X_project[1, :], 'ro')

plt.show()
```

```python
# Bonus: R and T calculation

K = np.array([[406.952636, 0.000000, 366.184147], [0.000000, 405.671292, 244.705127], [0.000000, 0.000000, 1]])
H = np.dot(np.linalg.pinv(K), H)

h1 = H[:,0]; h2 = H[:,1]; h3 = H[:,2]
A = np.zeros((3,3))
A[:,0] = h1; A[:,1] = h2; A[:,2] = np.cross(h1, h2)
# print(A)
(U, S, Vh) = np.linalg.svd(A, full_matrices=True)
R = np.dot(U, Vh)
t = h3/np.linalg.norm(h1)

print("Rotation matrix:\n", R)
print("Translation: \n", t)
```
